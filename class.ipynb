```python
# using constructor in class
class Employee: #defining class name
    def __init__(self, fname, lname, email, pay): 
    # __init__ method is the constructor
    # self is the automatic instance inside __init__ method of a class 
        self.name = fname
        self.lastname = lname
        self.connect = email
        self.salary = pay
    def email_id(self):
        return self.name + self.lastname + "@company.com"

emp_1 = Employee("Anil", "Jaiswal", "jaiswalaneel@gmail.com", 50000)
emp_2 = Employee("Sunil", "Jaiswal", "jaiswalsunil@gmail.com", 60000)
                #  these above are instances i.e objects of class employee
print(emp_1.connect)
print(emp_2.name,emp_2.lastname)
```

```python
emp_1.email_id() #calling with reference to constructor
```

```python
# copying in python
import copy  # we need to import the method first
list = ["a", "b", "c"]
#using shallow copy method
new_list = copy.copy(list)

print(list)

print(new_list)
```

```python
"""
Let's change the 1st element of the new_list and print both list
to conform that new_list is actually a copy to list rather than a reference.

"""
new_list[0] = 1
print(list)
print(new_list)
```

```python

anil=[[1,2,3],[4,5,6]]
sunil=copy.copy(anil)

print(anil)
print(sunil)
```

```python
"""But new_list still has reference to the each element of the object if 
the object is list or dictionary or set."""
sunil[0][2] = "Aj"  # it will change in both the nested lists
print(anil)
print(sunil)

# hence the name shallow copy
# it creates the copy of the object but not of the elements of objects
```

```python
# now let's dive into deepcopy
sushil = copy.deepcopy(anil)
sushil[0][2] = "SK"
print(anil)
print(sushil)
 
# Deep copy creates the copy of objects as well as of the elements of objects
```
